```python
import pandas as pd
from tqdm import tqdm
from ner_work.TopicWordRecognization import run_ner
test = pd.read_csv('./ner_work/test_public.csv')
model, tokenizer = run_ner.load_model('./ernie_ner_best.pdparams')
test['BIO_anno'] = None

for index, row in tqdm(test.iterrows()):
    input_text = row['text']
    ner_result = run_ner.cy_predict(model, tokenizer, input_text)
    #print('input_text:', input_text)
    #print('ner_result = ', ner_result)
    test.loc[index, 'BIO_anno'] = ' '.join(ner_result)

test
```

```python
# 数据保存
test['class'] = 0
test[['id', 'BIO_anno', 'class']].to_csv('./predict2-Ernie-aistudio.csv', index=False)
```
