# 数据集加载

```python
import pandas as pd

train = pd.read_csv('./train_data_public.csv')
train['BIO_anno'] = train['BIO_anno'].apply(lambda x: x.split(' '))
train['train_data'] = train.apply(lambda row: (list(row['text']), row['BIO_anno']), axis=1)
train
```

```python
test = pd.read_csv('test_public.csv')
test['test_data'] = test.apply(lambda row: (list(row['text'])), axis=1)
test
```

# 转换成BMES格式文件

```python
def tran2bmes(train, filename):
    file = open(filename, 'w', encoding='UTF-8')
    for index, row in train.iterrows():
        for ch, bio in zip(row['train_data'][0], row['BIO_anno']):
            # 空格
            ch = ch.strip()
            if ch != '' and ch != ' ' and ch != ' ':  # 英文空格和中文空格
                file.write("{} {}\n".format(ch, bio))
        file.write("\n")
    file.close()
    
# 生成训练集
tran2bmes(train, 'train.bmes')
```

```python
def tran2bmes_test(test, filename):
    file = open(filename, 'w', encoding='UTF-8')
    for index, row in test.iterrows():
        for ch in row['test_data']:
            # 空格
            ch = ch.strip()
            if ch != '' and ch != ' ' and ch != ' ':  # 英文空格和中文空格
                file.write("{} {}\n".format(ch, 'O'))
        file.write("\n")
    file.close()
    
# 生成测试集
tran2bmes_test(test, 'test.bmes')
```
