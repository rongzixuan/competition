# 数据预处理

```python
import pandas as pd

# 数据加载
train = pd.read_csv('./train.csv')
train
```

```python
test = pd.read_csv('./test.csv')
test
```

```python
# 合并train、test
data = pd.concat([train, test], axis=0)
data
```

```python
data.index = range(len(data))
data
```

```python
# 数据探索
data.isnull().sum()
```

```python
# 唯一值个数
for col in data.columns:
    print(col, data[col].nunique())
```

```python
# 去掉没有用的字段
data.drop('_c39',axis=1, inplace=True)
data
```

```python
# 类别特征
cat_columns = data.select_dtypes(include='O').columns  # 字母O，不是0
cat_columns
```

```python
column_name = []
unique_value = []
for col in cat_columns:
    print(col, data[col].nunique())
    column_name.append(col)
    unique_value.append(data[col].nunique())
    
df = pd.DataFrame()
df['col_name'] = column_name
df['value'] = unique_value
df = df.sort_values('value', ascending=False)
df
```

```python
data[cat_columns]
```

# 对包含？的字段进行处理

```python
# 单独看某个字段
data['property_damage'].value_counts()
data['property_damage'] = data['property_damage'].map({'NO': 0, 'YES': 1, '?': 2})
data['property_damage'].value_counts()
```

```python
data['police_report_available'].value_counts()
data['police_report_available'] = data['police_report_available'].map({'NO': 0, 'YES': 1, '?': 2})
data['police_report_available'].value_counts()
```

# 对日期特征进行处理

```python
# policy_bind_date、incident_date
data['policy_bind_date'] = pd.to_datetime(data['policy_bind_date'])
data['incident_date'] = pd.to_datetime(data['incident_date'])

# 查看最大日期、最小日期
data['policy_bind_date'].min()  # 1990-01-08 00:00:00
data['policy_bind_date'].max()  # 2015-02-22 00:00:00
data['incident_date'].min()     # 2015-01-01 00:00:00
data['incident_date'].max()     # 2015-03-01 00:00:00
```

```python
base_date = data['policy_bind_date'].min()
# 转换为date_diff
data['policy_bind_date_diff'] = (data['policy_bind_date'] - base_date).dt.days
```

```python
data['incident_date_diff'] = (data['incident_date'] - base_date).dt.days
data['incident_date_policy_bind_date_diff'] = (data['incident_date'] - data['policy_bind_date']).dt.days
data[['policy_bind_date', 'incident_date', 'policy_bind_date_diff', 'incident_date_diff', 'incident_date_policy_bind_date_diff']]
```

```python
data['policy_bind_date_weekday'] = data['policy_bind_date'].dt.weekday
data['incident_date_weekday'] = data['incident_date'].dt.weekday
data['policy_bind_date_weekday']
```

```python
# 去掉原始日期字段 policy_bind_date、incident_date
data.drop(['policy_bind_date', 'incident_date'], axis=1, inplace=True)
data
```

```python
data['policy_number'].value_counts()
```

```python
data.drop(['policy_number'],axis=1, inplace=True)
data
```

# 模型训练1

```python
train['fraud_reported'].mean()
```

```python
import xgboost as xgb
model_xgb = xgb.XGBClassifier(
            max_depth=6, learing_rate=0.05, n_estimators=2000,
            objective='binary:logistic', #tree_method='auto',
            subsample=0.8, colsample_bytree=0.8,
            min_child_samples=3, eval_metric='auc', reg_lambda=0.5
        )
model_xgb.fit(train.drop(['fraud_reported'], axis=1), train['fraud_reported'])
y_pred = model_xgb.predict_proba(test.drop(['fraud_reported'], axis=1))
y_pred
```

```python
y_pred[:, 1]
```

```python
result = pd.read_csv('./sampleSubmission.csv')
result['fraud_reported'] = y_pred[:, 1]
result.to_csv('./baseline3.csv', index=False)

# score = 0.85009
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,8))
plot_importance(model_xgb, max_num_features=30, ax=ax)
plt.show()
```
