# 数据集读取

```python
import pandas as pd

# 数据加载
train = pd.read_csv('./train.csv')
train
```

```python
test = pd.read_csv('./testA.csv')
test
```

# 数据预处理

```python
train['heartbeat_signals'][0].split(',')
```

```python
# 心跳信号一共205个时刻
len(train['heartbeat_signals'][0].split(','))
```

```python
columns = [x for x in range(205)]
signals_df = pd.DataFrame(train['heartbeat_signals'].str.split(',').tolist(), columns=(columns))
signals_df
```

```python
train1 = pd.merge(train, signals_df, left_index=True, right_index=True)
train1
```

```python
train1.drop(['heartbeat_signals'], axis=1, inplace=True)
train1
```

```python
import matplotlib.pyplot as plt

def show(index):
    tmp = train1.iloc[index]
    plt.plot(range(205), tmp[0: 205])
    plt.show()
    
for i in range(10):
    show(i)
```

```python
signals_df_test = pd.DataFrame(test['heartbeat_signals'].str.split(',').tolist(), columns=(columns))
signals_df_test
```

```python
test1 = pd.merge(test, signals_df_test, left_index=True, right_index=True)
test1
```

```python
test1.drop(['heartbeat_signals'], axis=1, inplace=True)
test1
```

```python
# 0-204时刻的类型为object，需要转换成float类型
train1.info()
```

```python
for col in range(205):
    train1[col] = train1[col].astype(float)
    test1[col] = test1[col].astype(float)
train1
```

```python
train1.to_pickle('./train1.pkl')
test1.to_pickle('./test1.pkl')
```

```python
train2 = train1.drop(['id', 'label'], axis=1)
test2 = test1.drop(['id'], axis=1)
train2
```

```python
train2 = train2.T
test2 = test2.T
train2
```

```python
tmp_train = train2.describe()
tmp_test = test2.describe()
tmp_train
```

```python
tmp_test
```

```python
tmp_train2 = tmp_train.T.drop(['count'], axis=1)
tmp_test2 = tmp_test.T.drop(['count'], axis=1)
tmp_train2
```

```python
train3 = pd.merge(train1, tmp_train2, left_index=True, right_index=True)
test3 = pd.merge(test1, tmp_test2, left_index=True, right_index=True)
train3
```

```python
test3
```

```python
test3.columns
```

```python
train3.to_pickle('./train3.pkl')
test3.to_pickle('./test3.pkl')
```

# 模型训练-lightgbm

```python
import lightgbm as lgb

clf = lgb.LGBMClassifier(
            num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='multiclass',
            max_depth=-1, learning_rate=0.01, min_child_samples=3, random_state=2022,
            n_estimators=1000, subsample=1, colsample_bytree=1,
        )

clf.fit(train3.drop(['id', 'label'], axis=1), train3['label'])
```

```python
y_pred = clf.predict_proba(test3.drop(['id'], axis=1))
```

```python
result = pd.read_csv('./sample_submit.csv')
result['label_0'] = y_pred[:, 0]
result['label_1'] = y_pred[:, 1]
result['label_2'] = y_pred[:, 2]
result['label_3'] = y_pred[:, 3]
result
```

```python
result.to_csv('./predict1-lightgbm1.csv', index=False)
```
