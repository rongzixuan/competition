特征提取-tsfresh

```python
%matplotlib inline
import matplotlib.pylab as plt

from tsfresh import extract_features, extract_relevant_features, select_features
from tsfresh.utilities.dataframe_functions import impute
from tsfresh.feature_extraction import ComprehensiveFCParameters, EfficientFCParameters

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 特征提取()
extraction_settings = ComprehensiveFCParameters()
```

```python
features_df = pd.DataFrame()
for i in range(0, train2['id'].max(), 1000):
    start_id = i
    end_id = i + 999
    print(start_id, end_id)
    # 切分数据集
    tmp_df = train2.iloc[start_id*205:  (end_id+1)*205]

    features = extract_features(tmp_df, column_id='id', column_sort='time',
                        default_fc_parameters = extraction_settings,
                        impute_function=impute)
    features_df = features_df.append(features, ignore_index=True)

features_df
```

```python
features_df.to_pickle('./features_df.pkl')

features_filtered = select_features(features_df, train2['label'])
features_filtered
```

```python
# 对test进行处理
features_test_df = pd.DataFrame()
for i in range(0, test2['id'].max(), 1000):
    start_id = i
    end_id = i + 999
    print(start_id, end_id)
    # 切分数据集
    tmp_df = test2.iloc[start_id*205:  (end_id+1)*205]

    features = extract_features(tmp_df, column_id='id', column_sort='time',
                        default_fc_parameters = extraction_settings,
                        impute_function=impute)
    features_test_df = features_test_df.append(features, ignore_index=True)

features_test_df
```

```python
features_test_df.to_pickle('./features_test_df.pkl')
```

```python
features_test_filtered = features_test_df[features_filtered.columns]
features_test_filtered
```
