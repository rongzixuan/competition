# 数据集加载

```python
import pandas as pd
import pickle

# 数据加载
with open('./train1.pkl', 'rb') as file:
    train1 = pickle.load(file)
train1
```

```python
# 数据加载
with open('./test1.pkl', 'rb') as file:
    test1 = pickle.load(file)
test1
```

# 模型训练-keras

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv1D, Conv2D, MaxPooling2D, Dense, Flatten, LeakyReLU, MaxPool1D, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
train_x = train1.drop(['id', 'label'], axis=1)
train_y = tf.keras.utils.to_categorical(train1['label'], 4)
train_x
```

```python
train_x = train_x.values.reshape(train_x.shape[0], 205, 1)
train_x
```

```python
train_y
```

```python
class CNNModel(tf.keras.Model):
    # 模型定义
    def __init__(self):
        super(CNNModel, self).__init__()           
        self.conv1 = Conv1D(filters=32, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001), input_shape = (205, 1))
        self.conv2 = Conv1D(filters=64, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001))
        self.conv3 = Conv1D(filters=128, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001))
        self.conv4 = Conv1D(filters=256, kernel_size=5, padding='same', activation=LeakyReLU(alpha=0.001))
        self.max_pool1 = MaxPool1D(pool_size=5, strides=2, padding='same')
        self.dropout = Dropout(0.5)
        self.flatten = Flatten()
        # 全连接层
        self.fc1 = Dense(units=512, activation=LeakyReLU(alpha=0.001))
        self.fc2 = Dense(units=4, activation='softmax')
            
    # 前向传播
    def call(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.max_pool1(x)
        x = self.dropout(x)
        x = self.flatten(x)
        # 全连接层
        x = self.fc1(x)
        x = self.fc2(x)
        return x
    
# 创建模型
model = CNNModel()
# 设置优化器
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['mae'])

# 传入训练数据进行训练
model.fit(train_x, train_y, batch_size=128, epochs=5, verbose=1)
```

```python
test_x = test1.drop(['id'], axis=1)
test_x = test_x.values.reshape(test_x.shape[0], 205, 1)
test_x
```

```python
y_pred = model.predict(test_x, batch_size=128)
y_pred
```

```python
result = pd.read_csv('./sample_submit.csv')

result['label_0'] = y_pred[:, 0]
result['label_1'] = y_pred[:, 1]
result['label_2'] = y_pred[:, 2]
result['label_3'] = y_pred[:, 3]
result
```

```python
result.to_csv('./predict3-keras-colab.csv', index=False)
```
