# 数据集加载

```python
def f(x):
    x1, x2, x3 = x.split('/')
    if x1 == '2022':
        x = str(int(x3) + 2000) + '/' + x2 + '/' + '1'
    return x
```

```python
import pandas as pd
# 数据加载
train_bank = pd.read_csv('./train_public2.csv')
train_bank['earlies_credit_mon'] = train_bank['earlies_credit_mon'].apply(f)
train_bank
```

```python
train_internet = pd.read_csv('./train_internet.csv')
train_internet
```

```python
test_public = pd.read_csv('./test_public2.csv')
test_public['earlies_credit_mon'] = test_public['earlies_credit_mon'].apply(f)
test_public
```

```python
train_bank.rename(columns={'isDefault': 'is_default'}, inplace=True)
train_bank
```

```python
# 查找数据集A、B之间共同的字段
common_cols = []
for col in train_bank.columns:
    if col in train_internet:
        common_cols.append(col)
common_cols
```

```python
len(common_cols)
```

```python
# 对于银行自己的数据集A，有3个不同的字段
train_bank_left = set(train_bank.columns) - set(common_cols)
print(train_bank_left)
# 对于网络公开的数据集B，有6个不同的字段
train_internet_left = set(train_internet.columns) - set(common_cols)
print(train_internet_left)
```

```python
# test有5000个数据，和bank字段一样 =》银行新产品，
# train_bank有1万数据（36个字段 + 3个不同） =》 银行新产品，样本少，想要训练一个精度高的模型
# train_internet有75万数据（共同36个字段 + 6个不同） =》别人的大量样本
# =》共有76万数据

# 获取数据集A、B中共同的部分
train1 = train_bank[common_cols]
print(train1)
train2 = train_internet[common_cols]
print(train2)
test = test_public[common_cols[:-1]]
print(test)
```

```python
train = pd.concat([train1, train2], axis=0)
train.index = range(len(train))
train
```

# 处理时间类型

```python
import datetime
train['issue_date'].dtype
```

```python
# 转换为pandas中的日期格式
train['issue_date'] = pd.to_datetime(train['issue_date'])
# 提取时间多尺度
train['issue_date_y'] = train['issue_date'].dt.year
train['issue_date_m'] = train['issue_date'].dt.month
#train['issue_date_d'] = train['issue_date'].dt.day
```

```python
# 转换为pandas中的日期格式
test['issue_date'] = pd.to_datetime(test['issue_date'])
# 提取时间多尺度
test['issue_date_y'] = test['issue_date'].dt.year
test['issue_date_m'] = test['issue_date'].dt.month
#train['issue_date_d'] = train['issue_date'].dt.day
```

```python
# 提取时间diff，最小日期是2007-06-01
base_time = min(train['issue_date'].min(), test['issue_date'].min())
base_time
```

```python
# train['issue_date_diff'] = train['issue_date'].apply(lambda x: x-base_time).dt.days
train['issue_date_diff'] = (train['issue_date'] - base_time).dt.days
test['issue_date_diff'] = (test['issue_date'] - base_time).dt.days
train['issue_date_diff']
```

```python
train.drop(['issue_date'], axis=1, inplace=True)
test.drop(['issue_date'], axis=1, inplace=True)
train
```

```python
# 转换为pandas中的日期格式
train['earlies_credit_mon'] = pd.to_datetime(train['earlies_credit_mon'])
# 提取时间多尺度
train['earlies_credit_mon_y'] = train['earlies_credit_mon'].dt.year
train['earlies_credit_mon_m'] = train['earlies_credit_mon'].dt.month
#train['issue_date_d'] = train['issue_date'].dt.day
```

```python
# 转换为pandas中的日期格式
test['earlies_credit_mon'] = pd.to_datetime(test['earlies_credit_mon'])
# 提取时间多尺度
test['earlies_credit_mon_y'] = test['earlies_credit_mon'].dt.year
test['earlies_credit_mon_m'] = test['earlies_credit_mon'].dt.month
#train['issue_date_d'] = train['issue_date'].dt.day
```

```python
# 提取时间diff，最小日期是1944-01-01
base_time2 = min(train['earlies_credit_mon'].min(), test['earlies_credit_mon'].min())
base_time2
```

```python
# train['earlies_credit_mon_diff'] = train['earlies_credit_mon'].apply(lambda x: x-base_time2).dt.days
train['earlies_credit_mon_diff'] = (train['earlies_credit_mon'] - base_time2).dt.days
test['earlies_credit_mon_diff'] = (test['earlies_credit_mon'] - base_time2).dt.days
train['earlies_credit_mon_diff']
```

```python
train.drop(['earlies_credit_mon'], axis=1, inplace=True)
test.drop(['earlies_credit_mon'], axis=1, inplace=True)
train
```
