```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

def transfer(x, kc_list):
    print('x:', x)
    if x == kc_list[0]:
        return 0
    elif x == kc_list[1]:
        return 1
    elif x == kc_list[2]:
        return 2
    elif x == kc_list[3]:
        return 3
    elif x == kc_list[4]:
        return 4

# 使用KMeans构造sub_classte特征
def KMeansCluster(data, class_label):    
    # 筛选class_label的数据
    data = data[data['class'] == class_label]
    # 数据归一化
    scaler = MinMaxScaler()
    scaler_data = scaler.fit_transform(data.drop(['loan_id', 'user_id', 'class', 'sub_class', 'is_default'], axis=1))
    # 使用Kmeans
    clf = KMeans(n_clusters=5, random_state=2022)
    pred = clf.fit(scaler_data)
    labels = pred.labels_
    print('labels:', labels)
    print('len(labels):', len(labels))
    kc = pred.cluster_centers_
    print('kc:', kc)
    print('kc[:, 0]:', kc[:, 0])
    print('kc[labels]:', kc[labels])
    kc_list = sorted([i for i in kc[:, 0]])
    print('kc_list:', kc_list)
    y = [transfer(x, kc_list) for x in kc[labels][:, 0]]
    #print("pred.labels_:", pred.labels_)
    print('y:', y)
    return y

# 构造sub_class
def sub_class_feature(train1, train2, test):
    data = pd.concat([train1, test])
    data = data.fillna(-1)
    data.index = range(len(data))
    s = []
    class_map = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E', 5: 'F', 6: 'G'}
    for class_ in range(7):
        # 针对class_label的数据，聚成5个子类
        data_label = KMeansCluster(data, class_)
        #label_map = {1: 1, 5: 2, 2: 3, 4: 4, 3: 5}      
        label = pd.Series(data_label, index=data.loc[data['class']==class_].index)
        #print('label:', label)
        #label = label.map(label_map)
        #print('label:', label)
        # 转换为英文类型
        class_tmp = class_map[class_]
        #print("class_tmp:", class_tmp)
        label2 = label.map({0: class_tmp+'1', 1: class_tmp+'2', 2: class_tmp+'3', 3: class_tmp+'4',
                                    4: class_tmp+'5'})
        #print('label2:', label2)
        s.append(label2)
    # 合并
    #train_sub_class = pd.concat(s).reset_index(drop=True)
    train_sub_class = pd.concat(s)
    return train_sub_class
        
lb = sub_class_feature(train_bank, train_internet, test)
data = pd.concat([train_bank, test])
data.index = range(len(data))
data['sub_class'] = lb
data[['class', 'sub_class']]
```

```python
df_train_bank = data[data['is_default'].notnull()]
df_train_bank.groupby(['sub_class'])['is_default'].mean().reset_index()
```
