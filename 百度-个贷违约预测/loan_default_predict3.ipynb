# 新特征构造

```python
# 1）信贷周转情况 monthly_payment_year_of_loan_recircle_b
# monthly_payment：分期付款金额
# year_of_loan：贷款年份
# recircle_b：信贷周转余额合计
train['monthly_payment_year_of_loan_recircle_b'] = train['monthly_payment'] * 12 * train['year_of_loan'] / (train['recircle_b'] + 0.1)
test['monthly_payment_year_of_loan_recircle_b'] = test['monthly_payment'] * 12 * test['year_of_loan'] / (test['recircle_b'] + 0.1)
train['monthly_payment_year_of_loan_recircle_b']
```

```python
# 2）单位循环额度利用率 recircle_u_recircle_b
# recircle_u：循环额度利用率
# recircle_b：信贷周转余额合计
train['recircle_u_recircle_b'] = train['recircle_u'] / (train['recircle_b'] + 0.1)
test['recircle_u_recircle_b'] = test['recircle_u'] / (test['recircle_b'] + 0.1)
train['recircle_u_recircle_b']
```

```python
# 3）债务收入比 debt_loan_ratio
# 对于 < 0 的异常值进行处理 =》 转换为正数
# 对于空缺的值进行处理 =》 fillna(1000)
train['debt_loan_ratio'] = np.abs(train['debt_loan_ratio']).fillna(1000)
test['debt_loan_ratio'] = np.abs(test['debt_loan_ratio']).fillna(1000)
train['debt_loan_ratio']
```

```python
# 4）信贷额度周转率 recircle_b_total_loan
train['recircle_b_total_loan'] = train['recircle_b'] / (train['total_loan'] + 0.1)
test['recircle_b_total_loan'] = test['recircle_b'] / (test['total_loan'] + 0.1)
train['recircle_b_total_loan']
```

```python
# 5）每期信贷额度周转情况 recircle_b_monthly_payment
# recircle_b：信贷周转余额合计
# monthly_payment：分期付款金额
train['recircle_b_monthly_payment'] = train['recircle_b'] / (train['monthly_payment'] + 0.1)
test['recircle_b_monthly_payment'] = test['recircle_b'] / (test['monthly_payment'] + 0.1)
train['recircle_b_monthly_payment']
```

```python
# 6）该笔订单空缺值情况
train['null_num'] = train.isnull().sum(axis=1)  # axis=1：该行的空缺值个数
test['null_num'] = test.isnull().sum(axis=1)
train['null_num']
```

```python
# 7）平均每年贷款金额 avg_total_loan
# total_loan：贷款数额
# year_of_loan：贷款年份
train['avg_total_loan'] = train['total_loan'] / train['year_of_loan']
test['avg_total_loan'] = test['total_loan'] / test['year_of_loan']
train['avg_total_loan']
```

```python
# 8）平均每年贷款利率 mean_interest
# interest：当前贷款利率
# year_of_loan：贷款年份
train['mean_interest'] = train['interest'] / train['year_of_loan']
test['mean_interest'] = test['interest'] / test['year_of_loan']
train['mean_interest']
```

```python
# 9）总计还款金额 all_monthly_payment
# monthly_payment：分期付款金额
# year_of_loan：贷款年份
train['all_monthly_payment'] = train['monthly_payment'] * train['year_of_loan']
test['all_monthly_payment'] = test['monthly_payment'] * test['year_of_loan']
train['all_monthly_payment']
```

```python
# 10）剩余公开记录的数量 reset_known_dero
# known_dero：贬损公共记录的数量
# pub_dero_bankrup：公开记录清除的数量
train['reset_known_dero'] = train['known_dero'] - train['pub_dero_bankrup']
test['reset_known_dero'] = test['known_dero'] - test['pub_dero_bankrup']
train['reset_known_dero']
```

```python
# 11）除信贷周转余额外，剩余贷款数额：rest_loan_recircle_b
# total_loan：贷款数额
# recircle_b：信贷周转余额合计
train['rest_loan_recircle_b'] = train['total_loan'] - train['recircle_b']
test['rest_loan_recircle_b'] = test['total_loan'] - test['recircle_b']
train['rest_loan_recircle_b']
```

```python
from autogluon.tabular import TabularPredictor

# 创建AutoML模型
model = TabularPredictor(label='is_default', path='./model', eval_metric='roc_auc')
# 可以指定训练时常
model.fit(train, time_limit=3000)
# 查看Leaderboard
model.leaderboard()
```

```python
# AutoML预测
y_pred = model.predict_proba(test)
y_pred
```
