# 识别

```python
from tools.infer import predict_rec
import os
from tqdm import tqdm
import pandas as pd
from tools.infer import utility

parser = utility.init_args()
parser
```

```python
args = parser.parse_args('')
args
```

```python
# 识别单张图片
path = '/home/aistudio/test_images'
result = pd.DataFrame(columns=['new_name', 'value'])
for filename in tqdm(os.listdir(path)):
    args.image_dir = '/home/aistudio/test_images/{}'.format(filename)
    args.rec_model_dir = '../ch_ppocr_server_v2.0_rec_infer'
    valid_image_file_list, rec_res = predict_rec.predict_ryt(args)
    tmp = {}
    tmp['new_name'] = filename
    tmp['value'] = rec_res
    result = result.append(tmp, ignore_index=True)

result
```

```python
# 识别整个文件夹
args.image_dir = '/home/aistudio/test_images'
args.rec_model_dir = '../ch_ppocr_server_v2.0_rec_infer'
args.batch_num = 64
valid_image_file_list, rec_res = predict_rec.predict_ryt(args)
rec_res
```

```python
len(rec_res)
```

```python
result = pd.DataFrame(columns=['new_name', 'value'])
for image_file, rec in zip(valid_image_file_list, rec_res):
    image_file = image_file.replace('/home/aistudio/test_images/', '')
    tmp = {}
    tmp['new_name'] = image_file
    tmp['value'] = rec[0]
    result = result.append(tmp, ignore_index=True)

result
```

```python
result.to_csv('./predict3-paddleOCR-aistudio.txt', sep='\t', index=False)
```
